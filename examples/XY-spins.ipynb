```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

import tqpt
```

Load the data.

```python
data = np.loadtxt('./data/XY.txt', skiprows=1, delimiter=',')
```

```python
X = data[:,1:]
y = data[:,0].astype('int')
```

```python
fs = 20
ms = 3

plt.scatter(X[:,0], X[:,-1], s=ms)
plt.xlabel(r'$\theta_{0}$', fontsize=fs)
plt.ylabel(r'$\theta_{99}$', fontsize=fs)
```

Perform KMeans on raw data.

```python
kmeans = KMeans(n_clusters=3)
raw_labels = kmeans.fit_predict(X)
```

```python
fs = 20
ms = 3

plt.scatter(X[:,0], X[:,-1], c=raw_labels, s=ms)
plt.xlabel(r'$\theta_{0}$', fontsize=fs)
plt.ylabel(r'$\theta_{99}$', fontsize=fs)
```

Transform the data via diffusion map.

```python
diffmap = tqpt.DCluster(X, metric='chebyshev', n_clusters=3, n_modes=5)

diffmap.solve(0.8)
```

```python
fs = 20
ms = 3

plt.scatter(X[:,0], X[:,-1], c=diffmap.labels_, s=ms)
plt.xlabel(r'$\theta_{0}$', fontsize=fs)
plt.ylabel(r'$\theta_{99}$', fontsize=fs)
```

Diffusion maps correctly clusters the data based on winding number.  True labels drawn below:

```python
fs = 20
ms = 3

plt.scatter(X[:,0], X[:,-1], c=y, s=ms)
plt.xlabel(r'$\theta_{0}$', fontsize=fs)
plt.ylabel(r'$\theta_{99}$', fontsize=fs)
```
